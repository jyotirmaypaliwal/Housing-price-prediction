# Housing price prediction using Linear Regression
## Here we are going to predict housing prices using Linear Regression and Sci-kit learn library. 

For more information on Linear Regression, please refer to https://machinelearningmastery.com/linear-regression-for-machine-learning/

Data was taken from https://www.kaggle.com/quantbruce/real-estate-price-prediction?ref=hackernoon.com




## Our first step will be to import various libraries.

```python
#Importing Libraries
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Loading data
### We have 5 main categories namely :
     - House age
     - Distance to the nearest MRT station
     - Number of convenience stores
     - Lattitude
     - Longitude
     - House price of unit area


```python
dataset = pd.read_csv("Real estate.csv")
dataset
```

## Visualizing data

### First we are going to plot the relationship between 'house age' and 'house price'

```python
house_age = dataset.iloc[:,1]
house_price = dataset.iloc[:,6]
plt.scatter(house_age, house_price, s=None, c='green', marker=None, cmap=None, vmin=None, vmax=None, alpha=None, linewidths=None, edgecolors='blue')
plt.title(' House age vs House price')
plt.xlabel('House age')
plt.ylabel('House price')
plt.show()
```

### Now we are going to plot the relationship between 'Number of convenience stores' and 'house price'

```python
convin_stores = dataset.iloc[:,3]
house_price = dataset.iloc[:,6]
plt.scatter(convin_stores, house_price, s=None, c='green', marker=None, cmap=None, vmin=None, vmax=None, alpha=None, linewidths=None, edgecolors='blue')
plt.title("#convenience stores vs House price")
plt.xlabel('No. of convenience stores')
plt.ylabel('House price')
plt.show()
```

## Data preprocessing

```python
X = dataset.iloc[:,1:6].values
y = dataset.iloc[:,6:7].values
```

### Shape

```python
X.shape
```

```python
y.shape
```

### Splitting data into train and test set 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

## Importing and instantiating model

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
```

## Training model

```python
lin_reg.fit(X_train,y_train)
```

## Evaluating model

```python
score_test = lin_reg.score(X_test, y_test)
print(score_test)
```

```python
print('Accuracy on the test set: ', score_test*100, '%')
```

```python
score_train = lin_reg.score(X_train, y_train)
print(score_train)
```

```python
print('Accuracy on the training set: ', score_train*100, '%')
```

```python
predictions = lin_reg.predict(X_test)
```

### Graph depicting our predictions vs actual values taking 'House age' as x-axis and 'House price' as y-axis.

```python
plt.plot(X_test[:,0], predictions, label='predictions')
plt.plot(X_test[:,0], y_test, label='actual')
plt.title('Predictions vs actual')
plt.legend()
```

### Graph depicting our predictions vs actual values taking 'No. of convenience stores' as x-axis and 'House price' as y-axis

```python
plt.plot(X_test[:,2], predictions, label='predictions')
plt.plot(X_test[:,2], y_test, label='actual')
plt.title('Predictions vs actual')
plt.legend()
```
